1933번 스카이라인 <span style="color:red">실패</span> - 2025.06.17

```python
import sys
import heapq
input = sys.stdin.readline

n = int(input())
buildings = []
height = [0 for _ in range(n)]
q = []

end = [0 for _ in range(n)]
check = set()
for i in range(n):
    l, h, r = map(int, input().split())

    buildings.append((l, i, 1)) # 시작점은 1
    buildings.append((r, i, -1)) # 끝점은 -1
    height[i] = h
    end[i] = r

# 1. 위치가 빠른 순
# 2. 위치가 같다면 시작점인지
# 3. 위치도 같고 시작점이라면 높이가 높은 순
buildings.sort(key=lambda x: (x[0], -x[2], -height[x[1]]))

max_height = 0
ans = []
for i in range(len(buildings)):
    cursor, idx, pt = buildings[i]

    if pt == 1:
        if max_height < height[idx]:
            max_height = height[idx]
            ans.append((cursor, max_height))
        heapq.heappush(q, (-height[idx], end[idx]))

    else:
        check.add(cursor)
        while q:
            if q[0][1] not in check:
                break
            heapq.heappop(q)

        if not q:
            if max_height:
                max_height = 0
                ans.append((cursor, max_height))
        else:
            if -q[0][0] != max_height:
                max_height = -q[0][0]
                ans.append((cursor, max_height))

for i in ans:
    print(i[0], i[1], end=' ')
```

```python
# 1. x 좌표를 기준으로 정렬
# 2. 각 점들을 이등분하면서 각 영역에 대해 가장 가까운 점을 찾음.
# 3. 왼쪽, 오른쪽 구역에 대해 가장 가까운 점을 찾은 뒤, 합침.
# 4. 합칠 때 중간 영역(가장 가까운 점의 거리안에 있는 (x좌표에서) 점들 중 더 가까운 점이 있는지 탐색)
# 5. 이때 브루트포스로 찾으면 최악일 때(가장 가까운 거리에 따라 다름.) 
#    O(N^2)이므로 y좌표를 기준으로 중간 영역에 있는 점에대해 다시 정렬 후 탐색. 이때 더 가까운 점이 있다면 break
# 6. 위 과정을 재귀로 반복하며 가장 가까운 점을 찾는다.
import sys
# input = sys.stdin.readline

def dist(x1, y1, x2, y2):
    return (x1 - x2) ** 2 + (y1 - y2) ** 2

def find_closest_dot(left, right):
    if left == right:
        return 1e10
    else:
        mid = (left + right)//2
        min_d = min(find_closest_dot(left, mid), find_closest_dot(mid + 1, right))
        middle_area = []

        for i in range(mid, left - 1, -1):
            if (dots[i][0] - dots[mid][0]) ** 2 < min_d:
                middle_area.append(dots[i])
            else:
                break
        for i in range(mid + 1, right + 1):
            if (dots[i][0] - dots[mid][0]) ** 2 < min_d:
                middle_area.append(dots[i])
            else:
                break
        
        middle_area.sort(key=lambda x: x[1])
        for i in range(len(middle_area) - 1):
            for j in range(i + 1, len(middle_area)):
                if (middle_area[i][1] - middle_area[j][1]) ** 2 < min_d:
                    d = dist(middle_area[i][0], middle_area[i][1], middle_area[j][0], middle_area[j][1])
                    min_d = min(min_d, d)
                else:
                    break
        return min_d

n = int(input())
dots = []
for _ in range(n):
    x, y = map(int, input().split())
    dots.append((x, y))
dots.sort()

if len(dots) != len(set(dots)):
    print(0)
else:
    print(find_closest_dot(0, len(dots) - 1))
```
