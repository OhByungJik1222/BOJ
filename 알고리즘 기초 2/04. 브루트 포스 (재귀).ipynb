14889번 스타트와 링크 <span style="color:green">성공</span> - 2025.03.21

```python
# 예시) n = 6
# 1. start 팀에 기본적으로 0을 넣어준다
# 123 124 125 126 134 135 136 145 146 156
# 456 356 346 345 256 246 245 236 235 234
# 이런식으로 1로 시작하는 조합만 뽑으면 나머지는 판별할 필요 X (중복됨)
# 2. 이후 dfs 함수의 depth를 1로 (start에 이미 값이 하나 들어 있으니까)
# 3. 이후 dfs 함수의 for문에서 조합 (6 - 1)C(n//2 - 1) 뽑는것과 똑같이 수행
# 4. depth가 n//2가 되었을때 집합 연산을 통해 원래 인덱스 {0, 1, 2, 3, 4, 5}에서 start 팀과의 차집합을 구해 link 팀에 저장해준다.
# 5. 이때 각각의 팀에서 2명씩 뽑아 능력치를 더해준다 (간단한 조합이라 for문으로 구현)
# 6. min() 함수로 값 비교후 ans에 저장

import sys

def dfs(start, depth):
    global ans

    if depth == n//2:
        start_stat = 0
        link_stat = 0
        link_team = list(set(num) - set(start_team)) # {0, 1, 2, 3, 4, 5} - {start_team 인덱스}
        
        for i in range(n//2 - 1): # 각 팀에서 두명을 뽑아 능력치 더하기
            for j in range(i + 1, n//2):
                start_stat += (team[start_team[i]][start_team[j]] + team[start_team[j]][start_team[i]])
                link_stat += (team[link_team[i]][link_team[j]] + team[link_team[j]][link_team[i]])
        ans = min(ans, abs(start_stat - link_stat)) # 비교

    for i in range(start, n - 1): # 조합 (6 - 1)C(n//2 - 1) 구하기
        start_team.append(num[i])
        dfs(i + 1, depth + 1)
        start_team.pop()

n = int(input()) # n == 6이면
num = [i for i in range(n)] # [0, 1, 2, 3, 4, 5]
team = [list(map(int, input().split())) for _ in range(n)]
ans = sys.maxsize
start_team = [num.pop(0)] # [0]

dfs(0, 1) # start_team에 0이 하나 이미 저장되어 있으므로

print(ans)

# 내가 짧게 작성해본 집합연산 예제
# a = list([1,2,3,4,5,6])
# b = list([4,5,6])

# a_s = set(a)
# b_s = set(b)

# print(a_s, b_s, a_s-b_s, list(a_s-b_s))
```
