1309번 동물원 <span style="color:green">성공</span> - 2025.03.10
 - <span style="color:green">25.03.14 복습</span>

```python
n = int(input())
dp = [[0, 0, 0] for _ in range(n + 1)]
dp[1] = [1, 1, 1]

for i in range(2, n + 1):
    dp[i][0] = sum(dp[i - 1])
    dp[i][1] = dp[i][0] - dp[i - 1][1]
    dp[i][2] = dp[i][0] - dp[i - 1][2]

print(sum(dp[n]))
```

17404번 RGB거리 2 <span style="color:red">실패</span> - 2025.03.04
 - 첫번째 색을 고정한 후, 합을 구하고 마지막 색의 값을 최대로 하여 선택 못하도록
 - <span style="color:green">25.03.14 복습</span>

 

```python
n = int(input())
a = [[0, 0, 0]] + [list(map(int, input().split())) for _ in range(n)]
dp = [a[i][:] for i in range(n + 1)]

ans = []
for i in range(3):
    dp[1] = [1001, 1001, 1001]
    dp[1][i] = a[1][i]
    dp[n][i] = 1001
    for j in range(1, n + 1):
        dp[j][0] += min(dp[j - 1][1], dp[j - 1][2])
        dp[j][1] += min(dp[j - 1][0], dp[j - 1][2])
        dp[j][2] += min(dp[j - 1][0], dp[j - 1][1])
    ans.append(min(dp[n]))
    dp = [a[i][:] for i in range(n + 1)]

print(min(ans))
```

2225번 합분해 <span style="color:green">성공</span> - 2025.03.10
 - <span style="color:green">25.03.14 복습</span>

```python
n, k = map(int, input().split())
dp = [[0 for _ in range(n + 1)] for _ in range(k + 1)]
for i in range(1, k + 1):
    dp[i][1] = i

for i in range(1, k + 1):
    for j in range(2, n + 1):
        dp[i][j] = (dp[i - 1][j] + dp[i][j - 1]) % 1000000000

print(dp[k][n])
```
