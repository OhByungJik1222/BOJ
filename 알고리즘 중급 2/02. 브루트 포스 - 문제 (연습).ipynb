```python
HORIZONTAL = 1
VERTICAL = 2
DIAGONAL = 3

def dfs(x, y, pipe):
    global ans

    if x == n - 1 and y == n - 1:
        ans += 1
        return

    if pipe == HORIZONTAL: # 가로
        put_horizontal(x, y)
        put_diagonal(x, y)

    elif pipe == VERTICAL: # 세로
        put_vertical(x, y)
        put_diagonal(x, y)

    elif pipe == DIAGONAL: # 대각
        put_horizontal(x, y)
        put_vertical(x, y)
        put_diagonal(x, y)

def put_horizontal(x, y):
    if 0 <= y + 1 < n:
        if graph[x][y + 1] == 0:
            graph[x][y + 1] = 1
            dfs(x, y + 1, HORIZONTAL)
            graph[x][y + 1] = 0

def put_vertical(x, y):
    if 0 <= x + 1 < n:
        if graph[x + 1][y] == 0:
            graph[x + 1][y] = 1
            dfs(x + 1, y, VERTICAL)
            graph[x + 1][y] = 0

def put_diagonal(x, y):
    if 0 <= x + 1 < n and 0 <= y + 1 < n:
        if graph[x][y + 1] == 0 and graph[x + 1][y + 1] == 0 and graph[x + 1][y] == 0:
            graph[x][y + 1] = 1
            graph[x + 1][y + 1] = 1
            graph[x + 1][y] = 1
            dfs(x + 1, y + 1, DIAGONAL)
            graph[x][y + 1] = 0
            graph[x + 1][y + 1] = 0
            graph[x + 1][y] = 0

n = int(input())
graph = [list(map(int, input().split())) for _ in range(n)]
graph[0][0], graph[0][1] = 1, 1

ans = 0
dfs(0, 1, HORIZONTAL)

print(ans)
```

```python
# 직전 파이프 옮기기 1 문제와 달리 n의 최대가 32이므로 단순 완탐으로는 해결 불가
# DP 사용해야함. 오랜만에 DP 문제라서 재밌었다.
n = int(input())
graph = [list(map(int, input().split())) for _ in range(n)]
graph[0][1] = 1

dp1 = [[0 for _ in range(n)] for _ in range(n)] # 가로
dp2 = [[0 for _ in range(n)] for _ in range(n)] # 세로
dp3 = [[0 for _ in range(n)] for _ in range(n)] # 대각
dp1[0][1] = 1
for i in range(n):
    for j in range(1, n):
        if graph[i][j] == 0:
            if 0 <= j - 1 < n:
                dp1[i][j] += dp1[i][j - 1]
                if 0 <= i - 1 < n:
                    dp1[i][j] += dp3[i][j - 1]

            if 0 <= i - 1 < n:
                dp2[i][j] += dp2[i - 1][j]
                if 0 <= j - 1 < n:
                    dp2[i][j] += dp3[i - 1][j]
            
            if 0 <= i - 1 < n and 0 <= j - 1 < n:
                if graph[i][j - 1] == 0 and graph[i - 1][j] == 0:
                    dp3[i][j] += dp1[i - 1][j - 1]  
                    dp3[i][j] += dp2[i - 1][j - 1]
                    dp3[i][j] += dp3[i - 1][j - 1]

print(dp1[n - 1][n - 1] + dp2[n - 1][n - 1] + dp3[n - 1][n - 1])
```
