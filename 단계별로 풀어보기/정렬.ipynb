 2751번 수 정렬하기2 <span style="color:red">오답</style>
  - 병합정렬, 퀵 정렬 알고리즘 사용한 코드 같이 보기 + input(), sys.stdin.readline() 차이 보기
  - 이 문제는 퀵 정렬 사용불가(O(n^2)), O(nlogn)만 사용가능(병합, 힙)

```python
import sys

n = int(input())
num = []

for i in range(n):
    num.append(int(sys.stdin.readline()))

num.sort()
for i in range(n):
    print(num[i])

'''
import sys

def Merge_Sort(arr): #병합정렬
    if len(arr) < 2: #리스트의 크기가 1이 될 때까지 쪼개기
        return arr
    mid = len(arr) // 2
    low_arr = Merge_Sort(arr[:mid])
    high_arr = Merge_Sort(arr[mid:])

    merged_arr = [] #다시 합칠 리스트 생성
    low = high = 0
    while low < len(low_arr) and high < len(high_arr): #다시 합치는 작업(크기를 비교하며)
        if low_arr[low] < high_arr[high]:
            merged_arr.append(low_arr[low])
            low += 1
        else:
            merged_arr.append(high_arr[high])
            high += 1
    merged_arr += low_arr[low:]
    merged_arr += high_arr[high:]
    return merged_arr

n = int(input())
num = []

for i in range(n):
    num.append(int(sys.stdin.readline()))

num = Merge_Sort(num)
for i in range(n):
    print(num[i])

--------------------------------------------------------------------------------------------------

import sys

def Quick_Sort(arr): #퀵정렬
    if len(arr) <= 1:
        return arr
    pivot = len(arr) // 2 #피봇(기준) 인덱스 저장
    front_arr, pivot_arr, back_arr = [], [], [] # 피봇 기준 작은 값, 큰 값, 같은 값 저장
    for value in arr:
        if value < arr[pivot]: #피봇 위치 값보다 작으면 작은 값 리스트에 저장
            front_arr.append(value)
        elif value > arr[pivot]: #피봇 위치 값보다 크면 큰 값 리스트에 저장
            back_arr.append(value)
        else: #피봇 위치 값과 같으면 같은 값 리스트에 저장장
            pivot_arr.append(value)
    return Quick_Sort(front_arr) + Quick_Sort(pivot_arr) + Quick_Sort(back_arr) #각 리스트 합치기

n = int(input())
num = []

for i in range(n):
    num.append(int(sys.stdin.readline()))

num = Quick_Sort(num)
for i in range(n):
    print(num[i])
'''
```

1181번 단어 정렬 <span style="color:blue">정답</style>
 - sort() 메서드와 달리 sorted() 함수는 어떤 입력값이든 받을 수 있음.
 - sort() 메서드는 따로 메모리 필요X (값을 반환하지 않고 입력함수 안에서 싱행되는 내장함수)
 - sorted() 함수는 값을 반환하기 때문에 원본 유지 필요할 떄 사용 (추가 메모리 발생)

```python
n = int(input())

words = []
for i in range(n):
    words.append(input())

words = list(set(words))
ans = []
for i in range(len(words)):
    ans.append((len(words[i]), words[i]))

ans = sorted(ans)
for i in range(len(words)):
    print(ans[i][1])
```
