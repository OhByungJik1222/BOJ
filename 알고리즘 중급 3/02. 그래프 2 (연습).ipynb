1948번 임계경로 <span style="color:red">실패</span> - 2025.11.21

```python
from collections import deque

n = int(input())
m = int(input())

graph = [[] for _ in range(n + 1)]
r_graph = [[] for _ in range(n + 1)]
level = [0 for _ in range(n + 1)]
for _ in range(m):
    s, e, t = map(int, input().split())
    graph[s].append([e, t])
    r_graph[e].append([s, t])
    level[e] += 1
start, end = map(int, input().split())

dist = [0 for _ in range(n + 1)]
q = deque()
q.append(start)
while q:
    curr = q.popleft()
    for next, time in graph[curr]:
        if dist[next] < dist[curr] + time:
            dist[next] = dist[curr] + time

        level[next] -= 1
        if level[next] == 0:
            q.append(next)

cnt = 0
visited = [0 for _ in range(n + 1)]
visited[end] = 1
q = deque()
q.append(end)
while q:
    curr = q.popleft()
    for prev, time in r_graph[curr]:
        if dist[prev] + time == dist[curr]:
            cnt += 1
            if not visited[prev]:
                visited[prev] = 1
                q.append(prev)

print(dist[end])
print(cnt)
```

```python
# 만약 일방통행 도로만으로 사이클이 형성된다면
# 양방통행 도로의 방향을 어떻게 설정하더라도 무조건 사이클이 형성됨 (일방통행 길은 수정불가!)
# 반대로 일방통행 도로만으로 사이클이 형성되지 않는다면
# 양방통행 도로를 위상정렬 순서에 따라 방향을 정하여 사이클이 형성되지 않도록 만들 수 있음
# 따라서 양방통행 도로를 배제하고 일방통행 도로만으로 사이클 여부 판단하면 되는 문제
from collections import deque

n = int(input())
graph = []
for _ in range(n):
    graph.append(list(input()))

edges = [[] for _ in range(n)]
level = [0 for _ in range(n)]
for i in range(n):
    for j in range(n):
        if graph[i][j] == 'Y' and graph[j][i] == 'N':
            edges[i].append(j)
            level[j] += 1

q = deque()
for i in range(n):
    if level[i] == 0:
        q.append(i)

while q:
    curr = q.popleft()

    for next in edges[curr]:
        level[next] -= 1
        if level[next] == 0:
            q.append(next)

for l in level:
    if l != 0:
        print('NO')
        break
else:
    print('YES')
```

1647번 도시 분할 계획 <span style="color:red">실패</span> - 2025.12.01

```python
# 최소 스패닝 트리를 구성한 뒤, 가장 큰 가중치를 가지는 간선을 지우면 되는 문제이다.
# import sys
# input = sys.stdin.readline

def find(x):
    if parent[x] != x:
        parent[x] = find(parent[x])
    return parent[x]

def union(a, b):
    a = find(a)
    b = find(b)

    if a < b:
        parent[b] = a
    else:
        parent[a] = b

n, m = map(int, input().split())

parent = [i for i in range(n + 1)]

edges = []
for _ in range(m):
    a, b, c = map(int, input().split())
    edges.append([c, a, b])

edges.sort()

result = 0
max_cost = 0
for edge in edges:
    cost, a, b = edge
    if find(a) != find(b):
        union(a, b)
        result += cost
        max_cost = max(max_cost, cost)
print(result - max_cost)
```
