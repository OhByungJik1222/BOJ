1948번 임계경로 <span style="color:red">실패</span> - 2025.11.21

```python
from collections import deque

n = int(input())
m = int(input())

graph = [[] for _ in range(n + 1)]
r_graph = [[] for _ in range(n + 1)]
level = [0 for _ in range(n + 1)]
for _ in range(m):
    s, e, t = map(int, input().split())
    graph[s].append([e, t])
    r_graph[e].append([s, t])
    level[e] += 1
start, end = map(int, input().split())

dist = [0 for _ in range(n + 1)]
q = deque()
q.append(start)
while q:
    curr = q.popleft()
    for next, time in graph[curr]:
        if dist[next] < dist[curr] + time:
            dist[next] = dist[curr] + time

        level[next] -= 1
        if level[next] == 0:
            q.append(next)

cnt = 0
visited = [0 for _ in range(n + 1)]
visited[end] = 1
q = deque()
q.append(end)
while q:
    curr = q.popleft()
    for prev, time in r_graph[curr]:
        if dist[prev] + time == dist[curr]:
            cnt += 1
            if not visited[prev]:
                visited[prev] = 1
                q.append(prev)

print(dist[end])
print(cnt)
```

1507번 궁금한 민호 <span style="color:red">실패</span> - 2025.11.25

```python
# 만약 graph[i][j] = graph[i][k] + graph[k][j] 라면
# i ←→ j 경로는 필요없는 경로이다.
import sys

n = int(input())
graph = []
for _ in range(n):
    graph.append(list(map(int, input().split())))

useless = [[0 for _ in range(n)] for _ in range(n)]
for k in range(n):
    for i in range(n):
        for j in range(n):
            if i == k or j == k or i == j:
                continue

            if graph[i][j] == graph[i][k] + graph[k][j]:
                useless[i][j] = 1
            elif graph[i][j] > graph[i][k] + graph[k][j]:
                print(-1)
                sys.exit()

ans = 0
for i in range(n):
    for j in range(i + 1, n):
        if not useless[i][j]:
            ans += graph[i][j]
print(ans)
```
