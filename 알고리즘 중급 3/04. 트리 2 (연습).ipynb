```python
# f(x) = "트리에서 x의 부모"
import sys
import math
# input = sys.stdin.readline

def find_n(n, x):
    curr = x
    for i in range(LOG + 1, -1, -1):
        if n & (1 << i):
            curr = parent[curr][i]
    return curr

n = int(input())
f = list(map(int, input().split()))

LOG = int(math.log2(500000))

parent = [[0 for _ in range(LOG + 1)] for _ in range(n + 1)]
for i in range(n):
    parent[i + 1][0] = f[i]

for k in range(1, LOG + 1):
    for i in range(1, n + 1):
        parent[i][k] = parent[parent[i][k - 1]][k - 1]

q = int(input())
for _ in range(q):
    n, x = map(int, input().split())
    print(find_n(n, x))
```
